```python
target = 'survived' if 'survived' in df.columns else 'Survived'
print("\nTarget counts:")
print(df[target].value_counts(dropna=False))
```

```python
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
```

```python
df_small = df[features + [target]].copy()
```

```python
df.head()
```

```python
df_small.head()
```

```python
df_small.shape
```

```python
from sklearn.impute import SimpleImputer
import numpy as np

X = df_small.drop(columns=[target]).copy()
y = df_small[target].astype(int).copy()

numeric_cols = ['Age', 'SibSp', 'Parch', 'Fare']
cat_cols = ['Sex', 'Pclass', 'Embarked']

# Show missing before
print("Missing before:")
print(X[numeric_cols + cat_cols].isna().sum())

# Numeric imputer
num_imputer = SimpleImputer(strategy='median')
X[numeric_cols] = num_imputer.fit_transform(X[numeric_cols])

# Categorical imputer
cat_imputer = SimpleImputer(strategy='most_frequent')
X[cat_cols] = cat_imputer.fit_transform(X[cat_cols])

print("\nMissing after:")
print(X[numeric_cols + cat_cols].isna().sum())
```

```python
X_encoded = pd.get_dummies(X, columns=cat_cols, drop_first=True)
print("Shape after encoding:", X_encoded.shape)
display(X_encoded.head())
```

```python
# STEP 7: Train-test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.25, random_state=42, stratify=y
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
# STEP 8: Scaling numeric features
from sklearn.preprocessing import StandardScaler

numeric_cols = ['Age', 'SibSp', 'Parch', 'Fare']  # same names as before
scaler = StandardScaler()

# Fit on train numeric columns, transform both
X_train[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_test[numeric_cols] = scaler.transform(X_test[numeric_cols])

# Show mean/std (approx 0 and 1 on train)
print("Train numeric mean (approx):", X_train[numeric_cols].mean().round(3).tolist())
print("Train numeric std (approx):", X_train[numeric_cols].std().round(3).tolist())
```

```python
# STEP 9: Train logistic regression
from sklearn.linear_model import LogisticRegression,

model = LogisticRegression()
model.fit(X_train, y_train)

print("Model trained. Coefficients shape:", model.coef_.shape)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report
import numpy as np

y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:,1]

acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc = roc_auc_score(y_test, y_proba)
cm = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {acc:.4f}")
print(f"Precision: {prec:.4f}")
print(f"Recall: {rec:.4f}")
print(f"F1-score: {f1:.4f}")
print(f"ROC AUC: {roc:.4f}")
print("\nConfusion Matrix:\n", cm)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
